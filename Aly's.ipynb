```python
ConditionGroupFileNames = os.listdir("/content/drive/MyDrive/depresjon/data/condition/")
ControlGroupFileNames = os.listdir("/content/drive/MyDrive/depresjon/data/control/")
```

```python
X = []
y = []

for fileName in ConditionGroupFileNames:
    df = pd.read_csv("/content/drive/MyDrive/depresjon/data/condition/"+str(fileName))
    dates = df['date'].unique()
    activityLevelsPerDay = []
    for date in dates:
        if len(df[df['date'] == date]) == 1440:
            temp = pd.DataFrame(df[df['date'] == date]).drop(columns=['timestamp','date'])
            activityLevelsPerDay.append(temp)
    for dailyActivityLevel in activityLevelsPerDay:
        activityVector = np.array(dailyActivityLevel["activity"])
        if len(activityVector) == 1440:
            X.append(activityVector)
            y.append(1)

```

```python

for fileName in ControlGroupFileNames:
    df = pd.read_csv("/content/drive/MyDrive/depresjon/data/control/"+str(fileName))
    dates = df['date'].unique()
    activityLevelsPerDay = []
    for date in dates:
        if len(df[df['date'] == date]) == 1440:
            temp = pd.DataFrame(df[df['date'] == date]).drop(columns=['timestamp','date'])
            activityLevelsPerDay.append(temp)
    for dailyActivityLevel in activityLevelsPerDay:
        activityVector = np.array(dailyActivityLevel["activity"])
        if len(activityVector) == 1440:
            X.append(activityVector)
            y.append(0)

```

```python
for i in range(0,len(X)):
    X[i] = [X[i]]
```
